```python
#Step2: Read data from the uploaded csv file
df = pd.read_csv('C:\\Users\\adevdas\\Downloads\\2049_m2_demo8_v1_plo_03quh0n\\2049_m2_demo8_v1.0\\cars.csv')
df.head()
```

```python
#Step4: Set up a heatmap using seaborn
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10,6))                        #figure size of (10,6)
sns.heatmap(df.corr(), center=0, cmap='Reds')
ax.set_title('Multi-Collinearity of Car Attributes')
```

This heatmap represents the collinearity of the multiple variables in the dataset. Since the model names are not numerical values, the unnamed column was dropped off.

Dark red represents a positive correlation, while light white is a negative correlation.

```python
# Step5: Annotate each cell with the numeric value
fig, ax = plt.subplots(figsize=(10,6))
sns.heatmap(df.corr(), center=0, cmap='BrBG', annot=True)      #adjusting the color and adding annotation (actual correlation values) makes it easier to form a conclusion 
```

# You can see that disp and cylinder are highly correlated (features with high correlation are more linearly dependent and hence have almost the same effect on the dependent variable mpg, that is, 0.85 and 0.85).

When two features have high correlation, we can drop one of the two features and still produce a fairly accurate prediction model.
