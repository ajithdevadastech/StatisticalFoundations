```python
#Step1: Import the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
#Step2: Load the built-in dataset
tips = sns.load_dataset("tips")
tips.head()
```

```python
#Step3: Fetch the preview of dataset
tips.head(10)
```

```python
#Step4: Obtain generic mathematical details
tips.describe()
```

```python
#Step5: Visualize basic scatter plot
sns.lmplot(x="total_bill", y="tip", data=tips)
```

# The linear line across the plot shows the trend of the tip given by the customers as per the total bill.

The data points farther away from the line (top-right) are the outliers or exceptions in the dataset.

The convergent point is the statistical mean or the generalized prediction of the tip value on a daily basis.

For sns.lmplot(), there are three mandatory parameters as illustrated above. There is also an optional hue parameter that takes in categorical columns and helps group the data plot as per hue parameter values.

```python
#Step6: Visualize basic scatter plot with hue parameter
sns.lmplot(x="total_bill", y="tip", hue="sex", data=tips)
```

# From the above plot, we can concur that while females are mostly sticking to the convergent point and tip the usual amount, quite a few males tip (exceptionally) more or less. So, hue helped us visualize this difference in separate color plotting.

```python
#Step7: Visualize basic scatter plot with marker parameter
sns.lmplot(x="total_bill", y="tip", hue="sex", data=tips, markers=["o","x"])
```

```python
#Step8: Visualization on separate facets
sns.lmplot(x="total_bill", y="tip", col="sex", data=tips)
```

```python
sns.lmplot(x="total_bill", y="tip", row="smoker", col="sex", data=tips)
```
