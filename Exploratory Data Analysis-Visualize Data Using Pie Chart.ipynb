# Exploratory Data Analysis
Demo 2: Visualize Data Using Pie Chart
In this demo, you will be shown how to visualize simple categorical data using a pie chart and its parameters.

```python
#Step1: Import the required library
import matplotlib.pyplot as plt
```

```python
#Step2: Enter the data
languages = 'Python', 'Java', 'PHP', 'JavaScript', 'C#', 'C++'
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
```

```python
#Step3: Assign the colors
colors = ['#FA8F0A','#F0DA32','#61F527','#18DE9D','#3594FF','#8B5A8C']
```

```python
#Step4: Plot the chart
plt.pie(popularity, labels=languages, colors=colors)    
plt.title('Popularity of Programming Languages in 2019')
plt.show()
```

##Let's look at a few more parameters of matplotlib.pyplot.pie:

```python
#Step5: autopct parameter
plt.pie(popularity, labels=languages, colors=colors, autopct='%1.1f%%')
plt.show()
```

```python
##autopct is a string or function used to label the wedges with their numeric value. The label is placed inside the wedge.
```

```python
#Step6: explode parameter
explode = (0.1, 0, 0, 0, 0, 0)
plt.pie(popularity, labels=languages, colors=colors, autopct='%1.1f%%', explode=explode)
plt.show()
```

```python
##explode is a len(x) array which specifies the fraction of the radius with which to offset each wedge
```

```python
#Step7: shadow parameter
plt.pie(popularity, labels=languages, colors=colors, autopct='%1.1f%%', explode=explode, shadow=True)
plt.show()
```

```python
##shadow parameter draws a shadow beneath the pie.
```

```python
#Step8: startangle parameter
plt.pie(popularity, labels=languages, colors=colors, autopct='%1.1f%%', explode=explode, shadow=True, startangle=140)
plt.show()
```

```python
##startangle rotates the start of the pie chart by angle degrees counterclockwise from the x-axis.
```

```python
#Step9: frame parameter
plt.pie(popularity, labels=languages, colors=colors, autopct='%1.1f%%', explode=explode, shadow=True, startangle=140,
        frame=True)
plt.show()
```

```python
##frame parameter plots axes frame with the chart if true.
```

```python
#Step10: rotatelabels parameter
plt.pie(popularity, labels=languages, colors=colors, autopct='%1.1f%%', explode=explode, shadow=True, startangle=140,
        frame=True, rotatelabels=True)
plt.show()
```

```python
##rotatelabels parameter rotates each label to the angle of the corresponding slice if true.
```

```python
#Step11: wedgeprops parameter
plt.pie(popularity, labels=languages, colors=colors, autopct='%1.1f%%', explode=explode, shadow=True, startangle=140,
        frame=True, rotatelabels=True, wedgeprops = {'linewidth': 3})
plt.show()
```

```python
##wedgeprops parameter sets the width of the wedge border lines.
```

```python
#Step12: radius parameter
plt.pie(popularity, labels=languages, colors=colors, autopct='%1.1f%%', explode=explode, shadow=True, startangle=140,
        frame=True, rotatelabels=True, wedgeprops = {'linewidth': 3}, radius=5)
plt.show()
```

```python
##The radius of the pie, if radius is None it will be set to 1.
```
